# Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using count vectorizer create bag of words

```python
cv = CountVectorizer(max_features=2500)
# Sorting independent and dependent variable 
X = cv.fit_transform(corpus).toarray()
Y = data['feedback'].values
```

```python
# Saving the Count Vectorizer
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

```python
X.shape
```

```python
Y.shape
```

```python
# Splitting data into train test and split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=3)

X_train.shape
```

```python
X_train.max()
```

```python
# Scale X_train and X_test so that values can be in between 0 and 1
Scaler = MinMaxScaler()
X_train_scl = Scaler.fit_transform(X_train)
X_test_scl = Scaler.fit_transform(X_test)
```

```python
# Saving the Scaller model
pickle.dump(Scaler, open('Models/scaler.pkl', 'wb'))
```

# Random Forest

```python
# Fitting scalled X_train and Y_train 
model_rf = RandomForestClassifier()
model_rf.fit(X_train_scl, Y_train)
```

```python
# Accuracy of model in training data
model_rf.score(X_train_scl, Y_train)
```

```python
# Accuracy of model in testing data
model_rf.score(X_test_scl, Y_test)
```

```python
# Predicting on the test set 
y_preds = model_rf.predict(X_test_scl)
```

```python
y_preds
```

```python
cm = confusion_matrix(Y_test, y_preds)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=model_rf.classes_)
cm_display.plot()
plt.show()
```

# K fold Cross Validation

```python
accuracies = cross_val_score(estimator=model_rf, X=X_train_scl, y=Y_train, cv=10)
accuracies.mean()
accuracies.std()
```

```python
# Applying Grid search CV to get optimal parameter 

params = {
    'bootstrap':[True],
    'max_depth':[80,100],
    'min_samples_split':[8,12],
    'n_estimators':[100,300]
}
```

```python
cv_object = StratifiedKFold(n_splits=2)
grid_search = GridSearchCV(estimator=model_rf, param_grid=params, cv= cv_object, verbose=0, return_train_score=True)
grid_search.fit(X_train_scl, Y_train.ravel())
```

```python
print(" Best parameter Combo : {}".format(grid_search.best_params_))
```

```python
print("Cross validation mean accuracy on train test : {}".format(grid_search.cv_results_['mean_train_score'].mean()*100))
print("Cross validation mean accuracy on test test : {}".format(grid_search.cv_results_['mean_test_score'].mean()*100))
print("Accuracy for test set : ", accuracy_score(Y_test,y_preds))
```
